```python
#importing header files

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

#Loading CSV file to dataframe df

df = pd.read_csv("C:\\MY STUFF\\Projects\\Invoice Management System\\1806597.csv")
df.head()
```

```python
#Printing df
df
```

# Preprocessing

```python
#Checking for Null Values
df.isna().sum()
```

```python
#Droping area_business because it contains all NULL values

df.drop(['area_business'], axis = 1, inplace = True)

#Printing df
df
```

```python
# Storing rows where clear_date is NULL into another df

df_NaN = df.loc[pd.isnull(df.clear_date)]
df_NaN.reset_index()
df_NaN
```

```python
#droping rows where clear_date is NULL
df.dropna(subset = ['clear_date'], axis = 0, inplace = True)
df
```

```python
#Checking for Unique Values
df.nunique()
```

```python
#Droping invoice_id & doc_id because it has unique values

df.drop(['invoice_id'], axis = 1, inplace = True)
df.drop(['doc_id'], axis = 1, inplace = True)
df
```

```python
#checking for no.of unique values in each column
df.nunique()
```

```python
#droping columns which are constant

df.drop(columns=df.columns[df.nunique()==1], inplace = True)
df
```

```python
#dropping the unnecessary columns

df.drop(['name_customer','buisness_year','document type'],axis=1,inplace=True)
df
```

```python
#converting column data to date time format

df['clear_date'] = pd.to_datetime(df['clear_date'])
df['posting_date'] = pd.to_datetime(df['posting_date'])
df['document_create_date'] = pd.to_datetime(df['document_create_date'],format="%Y%m%d")
df['document_create_date.1'] = pd.to_datetime(df['document_create_date.1'],format="%Y%m%d")
df['due_in_date'] = pd.to_datetime(df['due_in_date'],format="%Y%m%d")
df['baseline_create_date'] = pd.to_datetime(df['baseline_create_date'],format="%Y%m%d")
df
```

```python
#checking Data types of columns
df.dtypes
```

```python
#checking for duplicate rows


df.drop_duplicates(keep = 'first',inplace=True)
df
```

```python
#checking faulty datas
df[(df['document_create_date'] <= df['due_in_date'])]
```

```python
#checking for no.of unique values in each column

df.nunique()
```

```python
#checking unique values for invoice_currency

df.invoice_currency.unique()
```

```python
df.tail(20)
```

```python
#converting CAD to USD 

df['total_open_amount'].where(df['invoice_currency']=='CAD', 0.74*df['total_open_amount'], inplace = True)
df.tail(20)
```

```python
#sorting by document_create_date
df.sort_values(by = 'document_create_date', inplace = True)
df
```

```python
#dropping unnecessary columns
df.drop(['document_create_date.1','invoice_currency'], axis=1, inplace = True)
df
```

```python
#creating target column delay = clear date - due in date
df['delay'] = df['clear_date'] - df['due_in_date']
df
```

# Feature Engineering

```python
X_train.columns
```

```python
X_train.dtypes
```

```python
X_train.business_code.unique()
```

```python
#one hot encodding business_code

embarked_dummies = pd.get_dummies(X_train['business_code'], prefix='B_Code_')
X_train = pd.concat([X_train, embarked_dummies], axis=1)
X_train.drop('business_code', axis=1,inplace=True)
X_train.head()
```

```python
embarked_dummies = pd.get_dummies(X_val['business_code'], prefix='B_Code_')
X_val = pd.concat([X_val, embarked_dummies], axis=1)
X_val.drop('business_code', axis=1,inplace=True)
X_val.head()
```

```python
#for train set

#Target Encoding cust_number and cust_payment_terms

#adding target to train set for target encoding
X_train['delay'] = y_train.dt.days

target_encode_cust_number = X_train.groupby('cust_number')['delay'].mean()
X_train['encoded_cust_number'] = X_train['cust_number'].map(target_encode_cust_number)

target_encode_cust_payment_terms = X_train.groupby('cust_payment_terms')['delay'].mean()
X_train['encoded_cust_payment_terms'] = X_train['cust_payment_terms'].map(target_encode_cust_payment_terms)

# Removing target from train set

X_train.drop('delay',axis=1,inplace=True)
X_train
```

```python
#for validation set
#adding target to val set for target encoding
X_val['delay'] = y_val.dt.days

target_encode_cust_number = X_val.groupby('cust_number')['delay'].mean()
X_val['encoded_cust_number'] = X_val['cust_number'].map(target_encode_cust_number)

target_encode_cust_payment_terms = X_val.groupby('cust_payment_terms')['delay'].mean()
X_val['encoded_cust_payment_terms'] = X_val['cust_payment_terms'].map(target_encode_cust_payment_terms)

# Removing target from train set

X_val.drop('delay',axis=1,inplace=True)
X_val
```

```python
#for test set

#adding target to train set for target encoding
X_test['delay'] = y_test.dt.days

target_encode_cust_number = X_test.groupby('cust_number')['delay'].mean()
X_test['encoded_cust_number'] = X_test['cust_number'].map(target_encode_cust_number)

target_encode_cust_payment_terms = X_test.groupby('cust_payment_terms')['delay'].mean()
X_test['encoded_cust_payment_terms'] = X_test['cust_payment_terms'].map(target_encode_cust_payment_terms)

# Removing target from train set

X_test.drop('delay',axis=1,inplace=True)
X_test
```

```python
#making given days to pay for train set

X_train['given_days_to_pay'] = X_train['due_in_date'] - X_train['baseline_create_date']
X_train['given_days_to_pay'] = X_train['given_days_to_pay'].dt.days
X_train.head()
```

```python
#making given days to pay for validation set
X_val['given_days_to_pay'] = X_val['due_in_date'] - X_val['baseline_create_date']
X_val['given_days_to_pay'] = X_val['given_days_to_pay'].dt.days
X_val.head()
```

```python
#making given days to pay for test set

X_test['given_days_to_pay'] = X_test['due_in_date'] - X_test['baseline_create_date']
X_test['given_days_to_pay'] = X_test['given_days_to_pay'].dt.days
X_test.head()
```

```python
#Dropping object type columns
X_train.drop(['posting_date','document_create_date','cust_number','clear_date','due_in_date','baseline_create_date','cust_payment_terms'], axis=1, inplace=True)
X_train.dtypes
```
